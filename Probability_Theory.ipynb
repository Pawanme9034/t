# Counting Examples

```
def factorial(n):
  f = 1
  for i in range(1, n+1):
    f = f * i
  return f
```

```
def P(n, r):
  return int(factorial(n)/factorial(n-r))
```

```
def C(n, r):
  return int(factorial(n)/(factorial(r) * factorial(n-r)))
```
